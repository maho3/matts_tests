```python
diagpath = '/anvil/scratch/x-mho1/cmass-ili/mtnglike/fastpm/L3000-N384/11/diag/mtng_lightcone/hod00111_aug00000.h5'
with h5py.File(diagpath, 'r') as f:
    pkk = f['Pk_k3D'][:]
    pk = f['Pk'][:]
    bkk = f['Bk_k3D'][:]
    bk = f['Bk'][:]
```

```python
f, axs = plt.subplots(5, 1, figsize=(10, 20))

ax = axs[0]
ax.plot(pkk, pk[:, 0], label='Pk0')
ax.loglog()
ax.set(xlabel='k', ylabel='P0')

for i, ax in enumerate(axs[1:3]):
    ax.semilogx()
    ax.axhline(0, color='k', lw=0.5)
    ax.plot(pkk, pk[:, i+1]/pk[:, 0])
    ax.set(xlabel='k', ylabel=f'P{i+1}/P0')

ax = axs[3]
ax.axhline(0, color='k', lw=0.5)
ax.plot(bk[0], label='Bk0')
ax.set(xlabel='Bin Index', ylabel='B0')

ax = axs[4]
ax.axhline(0, color='k', lw=0.5)
ax.plot(bk[1]/bk[0], label='Bk1/Bk0')
ax.set(xlabel='Bin Index', ylabel='B1/B0')
```
