```python
f, axs = plt.subplots(7, 1, figsize=(8, 16))

ax = axs[0]
ax.plot(pkk, pk[:, 0], label='Pk0')
ax.loglog()
ax.set(xlabel='k', ylabel='P0')

for i, ax in enumerate(axs[1:3]):
    ax.semilogx()
    ax.axhline(0, color='k', lw=0.5)
    ax.plot(pkk, pk[:, i+1]/pk[:, 0])
    ax.set(xlabel='k', ylabel=f'P{i+1}/P0')

ax = axs[3]
ax.axhline(0, color='k', lw=0.5)
ax.plot(bk[0], label='k1k2k3Bk0')
ax.set(xlabel='Bin Index', ylabel='k1k2k3Bk0')

ax = axs[4]
ax.axhline(0, color='k', lw=0.5)
ax.plot(bk[1]/bk[0], label='Bk2/Bk0')
ax.set(xlabel='Bin Index', ylabel='B2/B0')

ax = axs[5]
ax.axhline(0, color='k', lw=0.5)
ax.plot(qk[0], label='k1k2k3Qk0')
ax.set(xlabel='Bin Index', ylabel='k1k2k3Qk0')

ax = axs[6]
ax.axhline(0, color='k', lw=0.5)
ax.plot(qk[1]/qk[0], label='Qk2/Qk0')
ax.set(xlabel='Bin Index', ylabel='Q2/Q0')
```
